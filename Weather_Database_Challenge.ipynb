```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy

# Import the datetime module from the datetime library.
from datetime import datetime

# import the request library. 
import requests

# import the API key
from config import weather_api_key
```

```python
# Step 3: Create a new set of 2,000 random latitudes and longitudes.

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Create a list for holding the cities.
cities = []

# Step 4: Get the nearest city using the citipy module. 
for coordinate in lat_lngs:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
```

```python
# Create an empty list to hold the weather data.
city_data = []

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

url = "http://api.openweathermap.org/data/2.5/weather?"
city_url = f"{url}appid={weather_api_key}&q="
scraped_API = requests.get(city_url+"Toronto").json()
scraped_API
```

```python
# Step 5: Perform an API call with the OpenWeatherMap.
# Step 6: Retrieve the following information from the API call: 
    # Latitude and longitude
    # Maximum temperature
    # Percent humidity
    # Percent cloudiness
    # Wind speed
    # Weather description (for example, clouds, fog, light rain, clear sky)
# Step 7: Add the data to a new DataFrame.


counter = 0 

for each_city in cities:
    city_url = f"{url}appid={weather_api_key}&q="
    final_url_results = requests.get(city_url+each_city).json()   
    
    try:
        
        weather_data = {
            "City Name": final_url_results['name'],
            "Latitude": final_url_results["coord"]['lat'],
            "Longitude": final_url_results["coord"]['lon'],
            "Max Temperature": final_url_results["main"]['temp_max'],
            "% Humidity":final_url_results["main"]['humidity'],
            "% Cloudiness":final_url_results["clouds"]["all"],
            "Wind Speed":final_url_results["wind"]["speed"],
            "Weather Description":final_url_results["weather"]["description"],
            
        }

        city_data.append(weather_data)
        counter = counter + 1

        print(f"{final_url_results['name']}")
        
    except:
        print(f"data not found {each_city}")
    pass
    
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {each_city}")
    
    # Add 1 to the record count.
    record_count += 1
    set_count += 1

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
weather_data = pd.DataFrame(city_data)
weather_data.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_Database.csv"

# Export the City_Data into a CSV.
weather_data.to_csv(output_data_file, index_label="City_ID")
```
